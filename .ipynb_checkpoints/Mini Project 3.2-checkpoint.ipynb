```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import metrics

import chardet
from wordcloud import WordCloud
from datetime import datetime
```

```python
# Loading the Dataset
labels=['target','ids','date','flag','user','text']
df = pd.read_csv('training.1600000.processed.noemoticon.csv',names=labels, encoding='Windows-1252')
```

```python
# Final Shape
print(df.shape)
```

```python
# Value counts for 'target'
tar_value = df['target'].value_counts()
print('Value Count for:\n')
print(tar_value)
```

```python
# Reducing the dataset to 20,000 rows maximum with equal amounts from 0 & 4
neg_rows = df[df['target'] == 0].head(10000)
pos_rows = df[df['target'] == 4].head(10000)

# Combine the two data
df_final = pd.concat([neg_rows, pos_rows], ignore_index=True)
df_final['target'].value_counts()
```

```python
df_final
```

# Visualising the Data

```python
# Bar plot of 'target' column
plt.figure(figsize=(10,6))
df_final['target'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Distribution of Targets')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Word Clouds
fig, axes = plt.subplots(1, 2, figsize=(12,6))

titles = ['Word Cloud for Negative Target', 'Word Cloud for Positive Target'] 
for i, target in enumerate([0, 4]):
    subset = df_final[df_final['target'] == target]
    text = ''.join(subset['text'])
    wordcloud = WordCloud(width=400, height=300, background_color='white').generate(text)
    axes[i].imshow(wordcloud, interpolation='bilinear')
    axes[i].set_title(titles[i])
    axes[i].axis('off')

plt.show()
```

```python
# Converting the timestamp to datetime format
df_final['date'] = pd.to_datetime(df['date'])

# Grouping by date and target
#df_final['date'] = df_final['date'].dt.date
target_counts = df_final.groupby([df_final['date'].dt.date, 'target']).size().unstack(fill_value=0)

# Plotting Target over time
plt.figure(figsize=(10,6))

# Plotting Negative target (Target = 0)
plt.plot(target_counts.index, target_counts[0], marker='o', linestyle='-', color='r', label='Negative Targets')

# Plotting Positive target (Target = 4)
plt.plot(target_counts.index, target_counts[4], marker='o', linestyle='-', color='g', label='Positive Targets')

plt.title('Target over Time')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# Unsupervised Learning

## Pipeline

### Dimensionality Reduction (PCA) & Clustering (K-Means)

```python
# Applying PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_final)

# Convert to dataframe
df_pca = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
```

## Clustering (K-Means) 

```python
# Apply K-Means clustering to the PCA Results
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(df_pca)

# Adding cluster labels to the PCA DataFrame
df_pca['Cluster'] = clusters
```

## Visualising the Unsupervised Learning data

```python
# Visualising
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_pca, palette='viridis')
plt.title('Clusters of Customers after PCA')
plt.show()
```
